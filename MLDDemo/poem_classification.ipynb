```python
import pandas as pd
traindf = pd.read_csv("poem_train.csv", encoding="utf-8")
traindf
```

```python
testdf = pd.read_csv("poem_test.csv", encoding="utf-8")
testdf
```

```python
import jieba
jieba.load_userdict("dict.txt.big")
```

```python
s = "平林漠漠煙如織，寒山一帶傷心碧。\r\n暝色入高樓，有人樓上愁。".replace("\r", "").replace("\n", "")
" ".join(jieba.cut(s))
```

```python
def poemflow(p):
    s = p.replace("\r", "").replace("\n", "")
    return " ".join(jieba.cut(s))
traindf["內容"] = traindf["內容"].apply(poemflow)
traindf    
```

```python
testdf["內容"] = testdf["內容"].apply(poemflow)
testdf
```

```python
#分類的時候 答案一定要轉成 0, 1, 2, 3, 4
trans = {"李白":0, "杜甫":1, "白居易":2}
traindf["作者"] = traindf["作者"].replace(trans)
testdf["作者"] = testdf["作者"].replace(trans)                            
traindf
```

```python
#指定參數 axis = 0 表示要刪除觀測值（row），指定參數 axis = 1 表示要刪除欄位（column）
traindf = traindf.drop(["詩名"], axis=1)
testdf = testdf.drop(["詩名"], axis=1)
traindf
```

```python
testdf
```

```python
#只有訓練才fit
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
train_bag = vec.fit_transform(traindf["內容"])
testdf_bag = vec.transform(testdf["內容"])
```

```python
print("有幾行:", len(vec.vocabulary_))
vec.vocabulary_
```

```python
print(train_bag)
```

```python
#做出公式
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(train_bag, traindf["作者"])
```

```python
#驗證一下公式好壞
pre = clf.predict(testdf_bag)
print("預測:", list(pre))
print("答案:", list(testdf["作者"]))
```

```python
from sklearn.metrics import accuracy_score
print("正確率:", accuracy_score(pre, testdf["作者"]))
```
